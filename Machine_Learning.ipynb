```python
#X = dataset.loc[: , dataset.columns != 'RainTomorrow'] #more efficient
X = dataset.iloc[ : , 11:21] #Features

Y = dataset['RainTomorrow'] #Target
```

```python
from sklearn.model_selection import train_test_split

#now we are defining data to test and train
#we have to take around 20% or less of data to test our efficiency 

X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size = 0.2, random_state = 0)
```

```python
#Training_Phase
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state = 0)

#feeding data
rf.fit(X_train,Y_train)



y_pred = rf.predict(X_test) #model predicted and y_test original data

```

```python
#ACCURACY
```

```python
from sklearn.metrics import accuracy_score #checking result

print (accuracy_score(Y_test,y_pred)) #1st we send the original values then pred values
```
